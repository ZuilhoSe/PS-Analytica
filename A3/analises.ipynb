## Limpeza e Análises Descritivas dos Dados

```python
# Checando a quantidade de linhas e colunas de cada dataframe
print(df_pib.shape)
print(df_adh.shape)
print(df_edu.shape)

# Checando a quantidade de valores nulos em cada dataframe
print(df_pib.isnull().sum())
print(df_adh.isnull().sum())
print(df_edu.isnull().sum())
```

Como podemos notar, a maioria dos dataframes não possuem valores nulos, por isso, a limpeza dos dados será feita apenas se necessário nas colunas que possuem valores nulos.

```python
# PIB
# Removendo linhas com valores nulos
df_pib = df_pib.dropna()

# Calculo da media de cada coluna agrupando por sigla da UF
df_pib_mean = df_pib.groupby('sigla_uf').mean()
df_pib_mean.head()
```

```python
# ploting bar chart for each column in 6 subplots
fig, ax = plt.subplots(3, 2, figsize=(10, 10))
ax = ax.ravel()
for i, column in enumerate(df_pib_mean.drop(columns=['ano', 'va']).columns):
    sns.barplot(x=df_pib_mean.index, y=df_pib_mean[column], ax=ax[i])
    ax[i].set_title(column)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=90)
fig.suptitle('PIB médio por estado', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# ADH
# Removendo linhas com valores nulos
df_adh = df_adh.dropna()

# Mantendo colunas para a análise
columns_to_keep = ['ano', 'sigla_uf', 'expectativa_vida', 'expectativa_anos_estudo', 
                   'taxa_freq_bruta_basico', 'taxa_freq_bruta_fundamental', 
                   'taxa_freq_bruta_medio', 'taxa_freq_bruta_superior', 'indice_gini', 
                   'prop_pobreza', 'renda_pc', 'indice_escolaridade', 
                   'indice_frequencia_escolar', 'idhm', 'idhm_e', 'idhm_r']

df_adhf = df_adh[columns_to_keep]

# Calculo da media de cada coluna agrupando por sigla da UF
df_adh_mean = df_adhf.groupby('sigla_uf').mean()
df_adh_mean.head()
```

```python
# Educação
# Removendo linhas com valores nulos
df_edu = df_edu.dropna()

# Mantendo colunas para a análise
columns_to_keep = ['ano', 'sigla_uf', 'localizacao', 'rede', 
                   'had_em', 'taxa_aprovacao_em', 'taxa_reprovacao_em', 
                   'taxa_abandono_em']

df_eduf = df_edu[columns_to_keep]

# Calculo da media de cada coluna agrupando por sigla da UF
df_edu_mean = df_eduf.groupby('sigla_uf').mean()
df_edu_mean.head()
```

## Análise de Dados

```python
# Relação do PIB com IDHM
# Merge dos dataframes mantendo os estados em comum
df_pib_adh = pd.merge(df_pib_mean, df_adh_mean, left_index=True, right_index=True)

# Plotando scatter plot
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(data=df_pib_adh, x='va', y='idhm_e', ax=ax)
for line in range(0, df_pib_adh.shape[0]):
     ax.text(df_pib_adh.va[line], df_pib_adh.idhm_e[line], df_pib_adh.index[line], horizontalalignment='left', size='small', color='black')
ax.set_xlabel('PIB')
ax.set_ylabel('IDHM')
fig.suptitle('Relação entre PIB e IDHM_e', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Realizando uma regressão linear para verificar a relação entre PIB e IDHM
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Separando as variáveis independentes e dependentes
X = df_pib_adh['va'].values.reshape(-1, 1)
y = df_pib_adh['idhm_e'].values

# Instanciando o modelo
model = LinearRegression()

# Treinando o modelo
model.fit(X, y)

# Plotando a linha de regressão
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(data=df_pib_adh, x='va', y='idhm_e', ax=ax)
ax.set_xlabel('PIB')
ax.set_ylabel('IDHM')
ax.plot(X, model.predict(X), color='red')
fig.suptitle('Relação entre PIB e idhm_e', fontsize=16)
plt.tight_layout()
plt.show()
```

Como podemos perceber, existe um outlier no PIB, que é o estado de São Paulo, por isso, é importante analisar os dados sem a presença desse estado.

```python
# Análise sem SP
# Removendo SP
df_pib_adh_no_sp = df_pib_adh.drop('SP')

# Separando as variáveis independentes e dependentes
X = df_pib_adh_no_sp['va'].values.reshape(-1, 1)
y = df_pib_adh_no_sp['idhm_e'].values

# Instanciando o modelo
model = LinearRegression()

# Treinando o modelo
model.fit(X, y)

# Plotando a linha de regressão
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(data=df_pib_adh_no_sp, x='va', y='idhm_e', ax=ax)
ax.set_xlabel('PIB')
ax.set_ylabel('idhm_e')
ax.plot(X, model.predict(X), color='red')
fig.suptitle('Relação entre PIB e idhm_e sem SP', fontsize=16)
plt.tight_layout()
plt.show()
```

    Como podemos perceber, existe uma relação entre o PIB e os indices educacionais, tendo muita dispersão ao redor da linha de regressão, mas mantendo a tendencia de que quanto maior o PIB, maior o indice educacional.

Vamos analisar também a relação enter renda per capita e os indices educacionais, focando nos anos de estudo.

```python
# Relação dos anos de estudo com a renda per capita
# Merge dos dataframes mantendo os estados em comum
df_adh_edu = pd.merge(df_adh_mean, df_edu_mean, left_index=True, right_index=True)

# Plotando scatter plot
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(data=df_adh_edu, x='expectativa_anos_estudo', y='renda_pc', ax=ax)
for line in range(0, df_adh_edu.shape[0]):
     ax.text(df_adh_edu.expectativa_anos_estudo[line], df_adh_edu.renda_pc[line], df_adh_edu.index[line], horizontalalignment='left', size='small', color='black')
ax.set_xlabel('Expectativa de anos de estudo')
ax.set_ylabel('Renda per capita')
fig.suptitle('Relação entre expectativa de anos de estudo e renda per capita', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Realizando uma regressão linear para verificar a relação entre expectativa de anos de estudo e renda per capita
# Separando as variáveis independentes e dependentes
X = df_adh_edu['expectativa_anos_estudo'].values.reshape(-1, 1)
y = df_adh_edu['renda_pc'].values

# Instanciando o modelo
model = LinearRegression()

# Treinando o modelo
model.fit(X, y)

# Plotando a linha de regressão
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(data=df_adh_edu, x='expectativa_anos_estudo', y='renda_pc', ax=ax)
ax.set_xlabel('Expectativa de anos de estudo')
ax.set_ylabel('Renda per capita')
ax.plot(X, model.predict(X), color='red')
fig.suptitle('Relação entre expectativa de anos de estudo e renda per capita', fontsize=16)
plt.tight_layout()
plt.show()
```

Aqui, aqui novamente percebermos uma relação linear entre os índices analisados, mas com uma baixa dispersão, indicando uma alta correlação entre as variáveis. Para vericar a relação entre as variáveis, vamos calcular a correlação entre elas.

```python
# Calculando a correlação entre as variáveis
correlation = df_adh_edu.drop(columns=['ano_x']).corr()
correlation
```

```python
# Plotando o heatmap
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm', ax=ax)
fig.suptitle('Correlação entre as variáveis', fontsize=16)
plt.tight_layout()
plt.show()
```

Como podemos perceber, existe uma alta correlação entre as variáveis, indicando que quanto maior a renda per capita, maior o indice educacional. E além disso, outra correlação alta é a com a formação no ensino superior, o que atende as expectativas.
